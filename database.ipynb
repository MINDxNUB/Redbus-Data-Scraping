```python
import mysql.connector
import pandas as pd
import numpy as np
```

```python
bus1=pd.read_csv("D:/CSV files redbus/ksrtc.csv")
bus2=pd.read_csv("D:/CSV files redbus/wbsrtc.csv")
bus3=pd.read_csv("D:/CSV files redbus/tgsrtc.csv")
bus4=pd.read_csv("D:/CSV files redbus/kctl.csv")
bus5=pd.read_csv("D:/CSV files redbus/rstc.csv")
bus6=pd.read_csv("D:/CSV files redbus/sbstc.csv")
bus7=pd.read_csv("D:/CSV files redbus/hrtc.csv")
bus8=pd.read_csv("D:/CSV files redbus/astc.csv")
bus9=pd.read_csv("D:/CSV files redbus/pepsu.csv")
bus10=pd.read_csv("D:/CSV files redbus/bsrtc.csv")

finalbus=pd.concat([bus1,bus2,bus3,bus4,bus5,bus6,bus7,bus8,bus9,bus10],ignore_index=True)
finalbus.head(10)
```

```python
#replacing nan values
#finalbus = finalbus.where(pd.notnull(finalbus),None)
finalbus['Seats_Available'] = finalbus['Seats_Available'].fillna(0)
finalbus['Start_time'] = finalbus['Start_time'].fillna('00:00')
finalbus['End_time'] = finalbus['End_time'].fillna('00:00')
finalbus['Bus_name'] = finalbus['Bus_name'].fillna('Unknown')
finalbus['Total_duration'] = finalbus['Total_duration'].fillna('Unknown')
finalbus['Bus_type'] = finalbus['Bus_type'].fillna('Unknown')  
```

```python
#removing inr and converting to float
finalbus["Price"]=finalbus["Price"].str.replace("INR","").astype(float)
finalbus['Price'] = finalbus['Price'].fillna(0)
```

```python
#removing no. of reviews from rating and converting it to float
finalbus["Ratings"]=finalbus["Ratings"].str.replace("New","")
finalbus["Ratings"]=finalbus["Ratings"].str.strip()
finalbus["Ratings"]=finalbus["Ratings"].str.split().str[0]
finalbus["Ratings"] = pd.to_numeric(finalbus["Ratings"], errors='coerce')
finalbus['Ratings'] = finalbus['Ratings'].apply(lambda x: 0 if pd.isna(x) else x)
```

```python
#removing seat availble and converting it to integer
finalbus['Seats_Available'] = finalbus['Seats_Available'].str.extract('(\d+)')
finalbus['Seats_Available'] = pd.to_numeric(finalbus['Seats_Available'], errors='coerce')
finalbus['Seats_Available'] = finalbus['Seats_Available'].fillna(0).astype(int)
```
